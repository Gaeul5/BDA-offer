## 태양광 탄소 배출량

```python
# 발전량 차이 계산
solar['발전량 차이'] = solar['예측값'] - solar['태양광전체']

# 탄소 배출량 감소 비율 계산
solar['탄소 배출량 감소 비율'] = (solar['발전량 차이'] / solar['태양광전체']) * 16.8
```

```python
solar = solar.drop(columns=[ '일조(hr)', '일사(MJ/m2)'])
```

```python
solar
```

```python
# 전체 실제 발전량의 총합
total_actual_generation = solar['태양광전체'].sum()

# 전체 예측된 발전량의 총합
total_predicted_generation = solar['예측값'].sum()

# 발전량 차이
generation_difference = total_predicted_generation - total_actual_generation

# 발전량 증가 비율
increase_ratio = generation_difference / total_actual_generation

# 탄소 배출량 감소 비율
carbon_emission_reduction_ratio = increase_ratio * 16.8

# 결과 출력
print(f"전체 실제 발전량: {total_actual_generation:.2f}")
print(f"전체 예측된 발전량: {total_predicted_generation:.2f}")
print(f"발전량 차이: {generation_difference:.2f}")
print(f"탄소 배출량 감소 비율: {carbon_emission_reduction_ratio:.2f}%")
```

## 태양광 정산단가

```python
money = pd.read_csv('정산단가.csv')
```

```python
money = money[['기간','신재생_태양','신재생_풍력']]
```

```python
money['일시']=money['기간']
```

```python
money = money[['일시','신재생_태양','신재생_풍력']]
```

```python
money
```

```python
# 데이터프레임 병합 (기간 기준)
merged_df = pd.merge(money, solar, on = "일시")

# 태양광 발전량에 따른 총 가격 계산 (실제값과 예측값 각각에 대해)
merged_df['실제값_태양_가격'] = merged_df['태양광전체'] * merged_df['신재생_태양']
merged_df['예측값_태양_가격'] = merged_df['예측값'] * merged_df['신재생_태양']

# 가격 차이 계산
merged_df['태양_가격 차이'] = merged_df['예측값_태양_가격'] - merged_df['실제값_태양_가격']

# 전체 실제 발전량과 예측 발전량의 총 가격 계산
total_actual_solar_price = merged_df['실제값_태양_가격'].sum()
total_predicted_solar_price = merged_df['예측값_태양_가격'].sum()
total_solar_price_difference = total_predicted_solar_price - total_actual_solar_price

# 결과 출력
print(merged_df[['일시', '태양광전체', '예측값', '실제값_태양_가격', '예측값_태양_가격', '태양_가격 차이']])
print(f"전체 실제 태양광 발전량 총 가격: {total_actual_solar_price:.2f}")
print(f"전체 예측된 태양광 발전량 총 가격: {total_predicted_solar_price:.2f}")
print(f"총 태양광 가격 차이: {total_solar_price_difference:.2f}")
```

```python
# '일시' 열을 datetime 형식으로 변환
merged_df['일시'] = pd.to_datetime(merged_df['일시'])

# 월별로 그룹화
merged_df['월'] = merged_df['일시'].dt.to_period('M')

# 월별 총 가격 계산
monthly_summary = merged_df.groupby('월').agg(
    실제값_태양_총가격=('실제값_태양_가격', 'sum'),
    예측값_태양_총가격=('예측값_태양_가격', 'sum')
).reset_index()

# 가격 차이 계산
monthly_summary['태양_가격_차이'] = monthly_summary['예측값_태양_총가격'] - monthly_summary['실제값_태양_총가격']
# 소수점 두 자리까지 표시 설정
pd.options.display.float_format = '{:.2f}'.format
```

## 풍력 정산단가

```python
# 데이터프레임 병합 (기간 기준)
merged_df = pd.merge(money, wind, on = "일시")

# 태양광 발전량에 따른 총 가격 계산 (실제값과 예측값 각각에 대해)
merged_df['실제값_풍력_가격'] = merged_df['실제값'] * merged_df['신재생_풍력']
merged_df['예측값_풍력_가격'] = merged_df['랜덤 포레스트 예측값'] * merged_df['신재생_풍력']

# 가격 차이 계산
merged_df['풍력_가격 차이'] = merged_df['예측값_풍력_가격'] - merged_df['실제값_풍력_가격']

# 전체 실제 발전량과 예측 발전량의 총 가격 계산
total_actual_wind_price = merged_df['실제값_풍력_가격'].sum()
total_predicted_wind_price = merged_df['예측값_풍력_가격'].sum()
total_wind_price_difference = total_predicted_solar_price - total_actual_solar_price

# 결과 출력
print(f"전체 실제 풍력 발전량 총 가격: {total_actual_wind_price:.2f}")
print(f"전체 예측된 풍력 발전량 총 가격: {total_predicted_wind_price:.2f}")
print(f"총 풍력 가격 차이: {total_wind_price_difference:.2f}")
```

```python
# '일시' 열을 datetime 형식으로 변환
merged_df['일시'] = pd.to_datetime(merged_df['일시'])

# 월별로 그룹화
merged_df['월'] = merged_df['일시'].dt.to_period('M')

# 월별 총 가격 계산
monthly_summary = merged_df.groupby('월').agg(
    실제값_풍력_총가격=('실제값_풍력_가격', 'sum'),
    예측값_풍력_총가격=('예측값_풍력_가격', 'sum')
).reset_index()

# 가격 차이 계산
monthly_summary['풍력_가격_차이'] = monthly_summary['예측값_풍력_총가격'] - monthly_summary['실제값_풍력_총가격']
# 소수점 두 자리까지 표시 설정
pd.options.display.float_format = '{:.2f}'.format
```
